```python
import pandas as pd
import feedparser
import requests
import re
#from urllib import urlparse
import urllib
from bs4 import BeautifulSoup

import nltk
from nltk.corpus import stopwords
from nltk import FreqDist
from nltk.tokenize import RegexpTokenizer
```

```python
#tokanizes Words in strings
def tok(txt,stop):
    tokenizer = RegexpTokenizer(r'\w+')
    #tokens = tokenizer.tokenize(txt)
    #text =  nltk.word_tokenize(txt)
    return([i for i in txt.lower().split() if i not in stop])
```

```python
#Counts most frequent words in string
def frq(words,anzahl):
    fdist = FreqDist(words)
    return(fdist.most_common(anzahl))
```

```python
#Stopwords plus Additional
stopp = set(stopwords.words('german'))
stopp.update(['dass',
              'dafür',
              '-',
              '–',
              'werden',
              'werden.',
              'werden,',
              'statt',
              'mehr',
              'deshalb',
              'ebenso',
              'bisher',
              'z.b.',
              'herr',
              'frau',
              'geehrter',
              'geehrte',
              'grüße',
              'grüßen',
              'freundliche',
              'freundlichen',
              'vielen',
              'dank',
              'müssen',
              'frage',
              'immer',
              'schon',
              'viele'
             ])
```

```python
#Media-Sources
Newspaper = {
			"spiegel" : "http://www.spiegel.de/schlagzeilen/index.rss",
			"sueddeutsche" : "http://rss.sueddeutsche.de/app/service/rss/alles/index.rss?output=rss",
			"welt" : "https://www.welt.de/feeds/latest.rss"
			}
```

```python
test = 'http://www.spiegel.de/schlagzeilen/index.rss'
```

```python
Title = []
Datum = []
Base = []
Summary = []
Link = []
Terms = []
```

```python
All_feeds = []

for key, value in Newspaper.items():
    feeds = feedparser.parse(value)
    articles = {key:feeds}
    All_feeds.append(feeds)
```

```python
print(All_feeds[1])
```

```python
for feeds in All_feeds:
    for f in feeds.entries:
        article_summary = f['summary']
        article_base = f['title_detail']['base']
        #article_newspaper = [feeds+len(article_summary)]
        article_link = f['link']
        article_title = f['title_detail']['value']
        article_terms = f['tags'][0]['term']
        article_publish = f['published']

        Summary.append(article_summary)
        Link.append(article_link)
        Title.append(article_title)
        Datum.append(article_publish)
        Base.append(article_base)
        if article_terms != None:
            Terms.append(article_terms)
        else:
            Terms.append('No Terms')
```

```python
df = pd.DataFrame({'Title': Title,
                             'Newspaper':Base,
                             'Date': Datum,
                             'Link': Link,
                             'Terms': Terms,
                             'Summary': Summary
                            })
```

```python
df['key'] = df['Title'].str[:5]+df['Date'].str[:5]+df['Terms'].str[:5]


#range(len(df))+
```
